```python
#需求 11-12点每分钟温度变化折线图  温度范围在15-18  （上海 北京）
#x-分钟 y - 温度
# 1.准备数据
import matplotlib.pylab as plt
%matplotlib inline
import random
x = range(60)
y_shanghai = [random.uniform(15,18) for i in x]
y_beijing = [random.uniform(15,18) for i in x]
#2 画布初始化
plt.figure(figsize=(20,8),dpi=120)

#3画图
plt.plot(x,y_shanghai)
#设置刻度
x_label = ["11点{}分".format(i) for i in x]
plt.xticks(x[::5],x_label[::5])
plt.yticks(range(0,35,5))
#添加网格
plt.grid(linestyle='--',alpha=0.5)
# 添加标题
plt.xlabel("时间段")
plt.ylabel("温度（摄氏度）")
plt.title("某城市11-12点时间段温度变化")
# 展示
plt.show()
```

```python
#需求再添加一个城市
#x-分钟 y - 温度
# 1.准备数据
import matplotlib.pylab as plt
%matplotlib inline
import random
x = range(60)
y_shanghai = [random.uniform(15,18) for i in x]
y_beijing = [random.uniform(20,25) for i in x]
#2 画布初始化
plt.figure(figsize=(20,8),dpi=120)

#3画图
plt.plot(x,y_shanghai,color='b',linestyle='-',label='上海')
plt.plot(x,y_beijing,color='g',linestyle='--',label='北京') 
#设置图例位置
#plt.legend(loc='0')
plt.legend(loc=3)
#设置刻度
x_label = ["11点{}分".format(i) for i in x]
plt.xticks(x[::5],x_label[::5])
plt.yticks(range(10,35,5))
#添加网格
plt.grid(linestyle='--',alpha=0.5)
# 添加标题
plt.xlabel("时间段")
plt.ylabel("温度（摄氏度）")
plt.title("上海北京11-12点时间段温度变化")
# 展示
plt.show()
```

```python
#需求再添加一个城市
#x-分钟 y - 温度
# 1.准备数据
import matplotlib.pylab as plt
%matplotlib inline
import random
x = range(60)
y_shanghai = [random.uniform(15,18) for i in x]
y_beijing = [random.uniform(20,25) for i in x]
#2 画布初始化
# plt.figure(figsize=(20,8),dpi=80)
figure,axes = plt.subplots(nrows=1,ncols=2,figsize=(20,8),dpi=80)

#3画图
axes[0].plot(x,y_shanghai,color='b',linestyle='-',label='上海')
axes[1].plot(x,y_beijing,color='g',linestyle='--',label='北京') 
#设置图例位置
#plt.legend(loc='0')
axes[0].legend(loc=3)
axes[1].legend(loc=3)
#设置刻度
x_label = ["11点{}分".format(i) for i in x]
axes[0].set_xticks(x[::5])
axes[0].set_xticklabels(x_label)
axes[0].set_yticks(range(10,35,5))

axes[1].set_xticks(x[::5])
axes[1].set_xticklabels(x_label)
axes[1].set_yticks(range(10,35,5))


#添加网格
axes[0].grid(linestyle='--',alpha=0.5)
axes[1].grid(linestyle='--',alpha=0.5)
# 添加标题
axes[0].set_xlabel("时间段")
axes[0].set_ylabel("温度（摄氏度）")
axes[0].set_title("上海11-12点时间段温度变化")

axes[1].set_xlabel("时间段")
axes[1].set_ylabel("温度（摄氏度）")
axes[1].set_title("北京11-12点时间段温度变化")
# 展示
plt.show()
```

# 画折线图

```python
import numpy as np
# 1.准备x,y
x = np.linspace(-1,1,10000)
```

```python
x
```

```python
y = 2*x*x
```

```python
import matplotlib.pylab as plt
%matplotlib inline

#2 画布初始化
plt.figure(figsize=(20,8),dpi=120)

#3画图
plt.plot(x,y,color='r',linestyle='-')

#设置图例位置
#plt.legend(loc='0')
plt.legend(loc=3)
#设置刻度

plt.xticks(range(-2,2))
plt.yticks(range(-2,2))
#添加网格
plt.grid()
plt.legend()
# 添加标题
plt.xlabel("x轴")
plt.ylabel("y轴")
plt.title("图")
# 展示
plt.show()
```

### 画饼图

```python
#需求 11-12点每分钟温度变化折线图  温度范围在15-18  （上海 北京）
#x-分钟 y - 温度
# 1.准备数据
import matplotlib.pylab as plt
%matplotlib inline
import random
```

```python
# 1.准备数据
moviename = ['流浪地球','疯狂的外星人','飞驰人生','新喜剧之王']
count = [5031,2030,3045,2105]
# 2.准备画布
plt.figure(figsize=(20,8),dpi=80)

# 3.开始画
plt.pie(count,labels=moviename,autopct="%1.2f%%",colors=['r','g','b','y'])
plt.axis('equal')
plt.legend()
# 4.显示
plt.show()
```
