```python
#1.导入pd的df
import pandas as pd
dataframe = pd.read_csv('data1.csv',sep=',')
```

```python
#2.数据预处理
#在对数据处理之前，需要对数据的缺失值和异常值情况进行了解。Python内有describe()函数，可以了解数据集的缺失值、均值和中位数等。将结果打印到csv文件
dataframe.describe().to_csv('dataframeDescribe.csv')
dataframe.describe()
```

```python
#去掉重复记录
dataframe.drop_duplicates(keep='first',inplace=True)
```

```python
#Party_Id	ymon 两列不是特征，可以删除
dataframe.drop(['Party_Id','ymon'],axis=1,inplace=True)
```

```python
# 检查数据中是否有缺失值
dataframe.isnull().any()
```

```python
dataframe.head()
```

```python
#发现insure_M_sum 无数据，acct_M_sum、pcntcm_bondttl_bal_mavg6 的数据均为0，可以删除 缺数据 删除该列
dataframe.drop(['insure_M_sum','acct_M_sum','pcntcm_bondttl_bal_mavg6'],axis=1,inplace=True)
```

```python
#数据格式转换-找出需要被处理的列  
columns_to_treat = []
for i in range(len(dataframe.dtypes)):
    dtype = dataframe.dtypes[i]
    if str(dtype) == 'object':
        columns_to_treat.append(i)
        
dict_objects = dict()
for i in columns_to_treat:
    column = dataframe.iloc[:,i]
    if i not in dict_objects:
        dict_objects[i] = []
    for j in column : 
        if j not in dict_objects[i]:
            dict_objects[i].append(j)
```

```python
#需要处理的列为14列，查看14列信息
columns_to_treat
#列内容
dict_objects
```

```python
#初始化字典
d={'Chrem_oth_bal':1,
  'Insure_Invest_bal':2,
  'Curr_Dpsit_Bal':3,
  'Fund_Stock_Bal':4,
  'Chrem_ltkx_bal':5,
  'Trdpty_Presv_Mgmt_Acct_Bal':6,
  'Time_Dpsit_Bal':7,
  'Fund_Bond_Bal':8,
  'Fund_Mix_Bal':9,
  'Vouch_Tbond_Bal':10,
  'Insure_prot_bal':11,
  'Bank_Card_Dpsit_Bal':12,
  'acct_nm_bal':13,
  'Agent_Solid_Gold_Bal':14,
  'Bank_Card_Od_Bal':15,
  'Fund_Curr_Bal':16,
  'Amort_Pay_Bal_ToCny':17,
  'Book_Entry_Tbond_Bal':18,
  'Fny_Bal_ToUsd':19
}
```

```python
''' 18max_prod_type	持有最多的产品种类 转换'''
for i in range(len(dataframe[dataframe.keys()[14]])):
    if dataframe['max_prod_type'][i] in d :
        dataframe.at[i,'max_prod_type'] = d[dataframe['max_prod_type'][i]]
dataframe['max_prod_type'] = dataframe['max_prod_type'].astype(int)
```

```python
#查看信息
dataframe.info()
```

```python
#构造新特征
dataframe['M']=dataframe['Ttl_asset_mavg6']
dataframe['C']=((dataframe['pcntcm_chrem_bal_mavg6']+dataframe['pcntcm_dpsit_bal_mavg6']+dataframe['pcntcm_fund_bal_mavg6']+dataframe['pcntcm_gold_bal_mavg6']+dataframe['pcntcm_insure_bal_mavg6']+dataframe['pcntcm_trdpty_bal_mavg6'])*dataframe['Ttl_eqty_mavg6'])/dataframe['Ttl_asset_mavg6']
dataframe['L']=dataframe['trans_bal_r']
dataframe['R']=dataframe['risktk_r']
dataframe['F']=dataframe['ttl_F_mavg6']
#dataframe['E']=dataframe['gini']
```

```python
df=dataframe[['M','C','L','R','F']]
```

```python
df.describe()
```

```python
#把 资产列里的异常值变形。最高取2000w
for i in range(len(df['M'])):
    if df.iloc[i,0]>=20000000 :
        df.iloc[i,0]=20000000
df.describe()
```

```python
#把 交易频次的异常值变形。最高取50
for i in range(len(df['F'])):
    if df.iloc[i,4]>=50 :
        df.iloc[i,4]=50
df.describe()
```

```python
#标准化处理
df1=(df-df.mean(axis=0))/(df.std(axis=0))
```

```python
#kmeans聚类方法-并将结果打印
from sklearn.cluster import KMeans
from sklearn import metrics

kmeans = KMeans(n_clusters = 5)
s=kmeans.fit(df1)

#预测分值
y_pred = kmeans.fit_predict(df1)
print(metrics.calinski_harabaz_score(df1, y_pred))

r1 = pd.Series(kmeans.labels_).value_counts()#统计各个类别的数目
r2 = pd.DataFrame(kmeans.cluster_centers_)#找出聚类中心
r = pd.concat([r2,r1],axis=1)#得到聚类中心对应的类别下的数目
r.columns = list(df.columns) + ['cnt'] #重命名表头
r
```

```python
#详细输出原始数据及其类别

result = pd.concat([df, pd.Series(kmeans.labels_, index = df.index)], axis = 1)  #详细
result.to_csv('result.csv')
```

```python
#kmeans聚类方法--聚类选择方法 df1
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
d=[]
for i in range(2,11):
    kmeans = KMeans(n_clusters = i)
    s=kmeans.fit(df1)
    d.append(kmeans.inertia_)
plt.plot(range(2,11),d,marker='o')
plt.xlabel('number of cluster')
plt.ylabel('distortions')
plt.show()
```
